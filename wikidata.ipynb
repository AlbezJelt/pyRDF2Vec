```python
from pyrdf2vec import RDF2VecTransformer
from pyrdf2vec.graphs import KG
from pyrdf2vec import walkers, samplers
from cachetools import LFUCache
```

```python
entities = [
  'http://www.wikidata.org/entity/Q38' # Italy 
]
```

```python
knowledge_graph = KG(
  "https://query.wikidata.org/sparql",
  query_string="query",
  skip_predicates={},
  literals=[
            [
             "http://www.w3.org/2004/02/skos/core#prefLabel"]
            ],
  mul_req=False # NB - true non compatibile con wikidata
)
```

```python
walker = walkers.RandomWalker(max_depth=3, max_walks=10, with_reverse=False, md5_bytes=None)
```

```python
transformer = RDF2VecTransformer(
  walkers=[walker],
  verbose=2
)
```

```python
walks = transformer.get_walks(knowledge_graph, entities)
```
